The next few blocks of code remove columns where a majority of the items are null. a majority in tis scenario is greater than 2/3.

```python
#Check and store the number of rows
NRows = df_stack.shape[0]
#Create a threshold
Threshold = int(2*NRows/3)
```

```python
#Removing Columns with more than 2/3 Nulls    
for columns in df_stack:  
    Column = df_stack[columns]
    NumNull = int(Column.isnull().sum())
    if NumNull > (Threshold):
        del df_stack[columns]

#Check shape - it looks like 17 columns were removed and the number of rows stayed consistent
df_stack.shape
```

Next, the nulls in the remaining columns need to be filled in. This will be completed in 2 steps. First, the categorical columns will have their nulls filled in with the most popular value. Then, numerical columns will have their nulls filled in with their average value.

```python
#Check the dataframe description to build a baseline
df_stack.describe()
```

```python
# Of the kept columns, the nulls need to be filled in. For categorical values, it makes sense to fill in with the Mode
#The loop below looks at every column, counts the number of rows, detremines the first non-null. Then, if it is a categorical
#column, it fills the nulls in with the mode value

for columns in df_stack:
    Column = df_stack[columns]
    NonNull=Column.loc[~Column.isnull()].iloc[0]
    if(isinstance(NonNull,str)):      
       Mode = Column.mode()[0]
       df_stack[columns]=df_stack[columns].fillna(Mode)
```

```python
# Of the kept columns, the nulls need to be filled in. For numerical values, it makes sense to fill in with the Mean. 
#The loop below looks at every column, counts the number of rows, detremines the first non-null. Then, if it is a numeric
#column, it fills the nulls in with the mean value

for columns in df_stack:   
    Column = df_stack[columns]
    NRows = (Column.count())
    NonNull = Column.loc[~Column.isnull()].iloc[0]
    if(isinstance(NonNull,numbers.Number)):
       mean = df_stack[columns].mean()
       df_stack[columns] = df_stack[columns].fillna(mean)
```
